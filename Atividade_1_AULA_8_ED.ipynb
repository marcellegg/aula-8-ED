```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

arestas = [
    (1, 2), (1, 5),
    (2, 3), (2, 5),
    (3, 4),
    (4, 5), (4, 6)
]
G.add_edges_from(arestas)

graus = dict(G.degree())
```

```python
print("Graus de cada vértice:")
for vertice, grau in sorted(graus.items()):
    print(f"Vértice {vertice}: Grau {grau}")
```

```python
maior_grau = max(graus.values())
menor_grau = min(graus.values())

vertices_maior = [v for v, g in graus.items() if g == maior_grau]
vertices_menor = [v for v, g in graus.items() if g == menor_grau]
```

```python
print(f"\nMaior Grau: {maior_grau} (Pertence aos vértices: {vertices_maior})")
print(f"Menor Grau: {menor_grau} (Pertence ao vértice: {vertices_menor})")
```

```python
plt.figure(figsize=(6, 4))
nx.draw(G, with_labels=True, node_color='lightpink', node_size=1000, font_weight='bold', font_color='black', edge_color='gray')
plt.title("Graus de vértices > & <")
plt.show()
```

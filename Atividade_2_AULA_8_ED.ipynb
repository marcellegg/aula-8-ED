```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

arestas_direcionadas = [
    (1, 2), (1, 3),
    (3, 2), (3, 4),
    (4, 3)
]
G.add_edges_from(arestas_direcionadas)
```

```python
for vertice in sorted(G.nodes()):
    grau_saida = G.out_degree(vertice)
    grau_entrada = G.in_degree(vertice)
    print(f"Vértice {vertice}:")
    print(f"  Grau de Saída   d+({vertice}) = {grau_saida}")
    print(f"  Grau de Entrada d-({vertice}) = {grau_entrada}\n")
```

```python
plt.figure(figsize=(6, 4))
posicoes = nx.spring_layout(G, seed=42)
nx.draw(G, posicoes, with_labels=True, node_color='red', node_size=1200,
        font_weight='bold', font_color='black', edge_color='gray',
        arrows=True, arrowsize=20, width=2)

plt.title("Rede de Vértices Direcionada (Dígrafo)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo orientadoo
G = nx.DiGraph()

# Adicionar vértices
G.add_nodes_from(['Elvis Presley', 'Fã-1', 'Fã-2', 'Fã-3', 'Não-Fã'])
```

```python
G.add_edges_from([
    ('Fã-1','Elvis Presley'),
    ('Fã-2','Elvis Presley'),
    ('Fã-3', 'Elvis Presley'),])
```

```python
print(f"Nós do grafo: {G.nodes()}")
print(f"Arestas do grafro: {G.edges()}")
print(f"Contando o num de Nós: {G.number_of_nodes()}")
print(f"Contando o num de Arestas: {G.number_of_edges()}")
```

```python
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels= True, node_color='yellow',
        node_size=2800, font_color='black', font_size=15, arrowsize=15)

plt.title('Representação em Grafo simples')
plt.show()
```
